**STEP 2: Generate List**

```python
!python -m SigMamba_V1.utils.generate_lists \
--features "features/siglip2" \
--train_csv "SigMamba/sigmamba_release/examples/train.csv" \
--temporal_anno "SigMamba/sigmamba_release/examples/Temporal_Anomaly_Annotation_Test.txt" \
--train_output "train_list.txt" \
--test_output "test_list.txt"
```

**STEP 3: Verify Feature Size**

```python
from transformers import AutoConfig
config = AutoConfig.from_pretrained("google/siglip2-large-patch16-384")
print(config.vision_config.hidden_size)
```

**STEP 4: Train The Model**

You can use your own dataset features.

## OR

1. [VINAY-UMRETHE/Dataset-SigMamba-V1-Small](https://huggingface.co/datasets/VINAY-UMRETHE/Dataset-SigMamba-V1-Small)
2. [VINAY-UMRETHE/Dataset-SigMamba-V1-Large](https://huggingface.co/datasets/VINAY-UMRETHE/Dataset-SigMamba-V1-Large)

**NOTE:** The vision encoder used in the dataset should be used in training / export.

```python
!accelerate launch -m SigMamba_V1.sigmamba.train \
--features_path "features/siglip2" \
--vision_model "google/siglip2-large-patch16-384" \
--train_anno "train_list.txt" \
--test_anno "test_list.txt" \
--feature_size 1024 \
--seg_num 32 \
--model_dim 768 \
--model_depth 8 \
--batch_size 32 \
--save_steps 500 \
--output_dir "results/sigmamba_v1" \
--save_models "models/sigmamba_v1" \
--num_epochs 2000
```

**STEP 5: Export To HuggingFace**

```python
!python -m SigMamba_V1.utils.export \
--checkpoint "models/checkpoint-best" \
--output "sigmamba_exported" \
--vision_model "google/siglip2-large-patch16-384"
```

```python
from huggingface_hub import HfApi

api = HfApi()

api.upload_folder(
    folder_path="sigmamba_exported",
    repo_id="YOUR_NAME/SigMamba-V1-Trained",
    repo_type="model"
)
```
